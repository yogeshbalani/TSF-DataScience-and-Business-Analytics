<h4>from the above diagram, we can clearly see that company earns the most profit from California and Least profit from North Dakota, and the company have no profit from West Virginia and Wyoming so we need to look at that and make sure to earn profit from these states</h4>

```python
# Top 10 Profitable cities

dftop10_cities = df.groupby('City')['Profit'].sum().reset_index().sort_values(by='Profit',ascending=False)
dftop10_cities = dftop10_cities.head(10)
dftop10_cities
```

```python
#Visualizing the Top 10 Profitable Products

sns.catplot(x='City',y='Profit',data=dftop10_cities,kind='bar',aspect=1.5,height=9,palette='cubehelix_r')
plt.title('Top 10 profitable Cities',size=20) #'ch:2.5,-.2,dark=.3'
plt.xticks(size=15)
plt.yticks(size=15)
plt.ylabel('Comulative profit',size=18)
plt.xlabel('Cities',size=18)
plt.show()


# Comulative Profit : CP is the total net profits from the specified segments added together over a specific time Frame.
```

```python
# Top Profitable States

dftop10_States = df.groupby('State')['Profit'].sum().reset_index().sort_values(by='Profit',ascending=False)
dftop10_States = dftop10_States.head(10)
dftop10_States
```

```python
#Visualizing the Top 10 Profitable Products

sns.catplot(x='State',y='Profit',data=dftop10_States,kind='bar',aspect=1.5,height=9,palette= 'gist_earth')
plt.title('Top 10 profitable States',size=20) #'ch:2.5,-.2,dark=.3'
plt.xticks(size=15)
plt.yticks(size=15)
plt.ylabel('Cumulative profit',size=18)
plt.xlabel('States',size=18)

plt.show()
```

```python
dftop_Region = df.groupby('Region')['Profit'].sum().reset_index().sort_values(by='Profit',ascending=False)
dftop_Region = dftop_Region.head(10)
dftop_Region
```

```python
#Visualizing the Top Profitable Region

sns.catplot(x='Region',y='Profit',data=dftop_Region,kind='bar',aspect=1.5,height=9,palette= 'cool')
plt.title('Top profitable Regions',size=20) #'ch:2.5,-.2,dark=.3'
plt.xticks(size=15)
plt.yticks(size=15)
plt.ylabel('Cumulative profit',size=18)
plt.xlabel('Region',size=18)

plt.show()
```
